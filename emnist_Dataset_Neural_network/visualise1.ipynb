```python
from __future__ import absolute_import, division, print_function, unicode_literals
import time
# TensorFlow and` tf.keras
strt1time = time.time()
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import os
# print(tf.__version__)

fashion_mnist = keras.datasets.fashion_mnist

# print(help(fashion_mnist))
# print(fashion_mnist.load_data())
(train_image , trina_labels ),(test_images,test_lables) = fashion_mnist.load_data()
print(type(train_image))
print(train_image.shape)
```

```python
for i in train_image[1:4]:
    print(i.shape)
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
import random
```

```python
plt.figure()
plt.imshow(train_image[-6])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
train_image = train_image/255.0

test_images = test_images/250.0
```

```python

plt.figure(figsize=(10,10))
for iq in range(25):
    plt.subplot(5,5,iq+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    # plt.imshow(train_image[iq],cmap=plt.cm.binary)
    # plt.imshow(train_image[iq])
    plt.imshow(test_images[iq])


    plt.xlabel(class_names[trina_labels[iq]])
```

```python
mode1l = keras.Sequential([keras.layers.Flatten(input_shape=(28, 28)),
                           keras.layers.Dense(128, activation='relu'),
                           keras.layers.Dense(10,activation = 'softmax')])
```

```python
mode2l = keras.Sequential([keras.layers.Flatten(input_shape=(28, 28)),
                           keras.layers.Dense(128, activation='relu'),
                           keras.layers.Dense(10,activation = 'softmax')])
```

```python
mode1l.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
mode2l.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
mode1l.fit(train_image,trina_labels,epochs=10)
```

```python
testloss,test_acc = mode1l.evaluate(test_images,test_lables) #verboae will shwo you details

print("\n Accuracy ",test_acc)
```

```python
mode2l.fit(train_image,trina_labels,epochs=10)
```

```python
testloss2,test_acc2 = mode2l.evaluate(test_images,test_lables) #verboae will shwo you details

print("\n Accuracy ",test_acc2)
```

```python
predircion = mode1l.predict(test_images)
```
